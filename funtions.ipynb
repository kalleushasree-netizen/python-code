```python
def greater(n):
    return n > 15
num = [4,6,78,45,23]
print(list(filter(greater, num)))
```

```python
from functools import reduce
def maximum(a, b):
    if a > b:
        return a
    else:
        return b
num = [7,19,62,34]
print(reduce(maximum, num))
```

```python
def index(n):
    for i in range(len(n)):
        if n[i] > 50:
            return i
n = [6,78,59,23]
print(index(n))
```

```python
def filter_out_neg(n):
    return n>=0
num = [-5,7,0,2,-45]
print(list(filter(filter_out_neg, num)))
```

```python
def filter_out_positive(n):
    return n<=0
num = [8,0,-6,-1]
print(list(filter(filter_out_positive, num)))
```

```python
from functools import reduce
def even(x):
    return x % 2 == 0 
def mul(a, b):
    return a*b
num = [8,3,9,1,5]
even = list(filter(even, num))
print(reduce(mul, even))
```

```python
n = int(input("Enter a number: "))
if n <= 1:
    print("Not Prime")
elif n == 2 or n == 3 :
    print("Prime")
elif n % 2 == 0 or n % 3 == 0 :
    print("Not Prime")
else:
    print("Prime")
```
