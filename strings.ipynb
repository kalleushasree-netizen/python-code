```python
def capital(str):
    for i in str:
        print(i.capitalize())
words = ['she','is','beautiful']
capital(words)
```

```python
def palindrome(s):
    return s == s[::-1]
word = ['10066001','pipe','one','python']
print(list(filter(palindrome, word)))
```

```python
def concatenate(s):
    return s
word = ['she','is','beautiful']
print(" ".join(words))
```

```python
def long(s):
    return len(s) == max
word =  ['sugar','milk','curd']
max = 0
for i in word:
    if len(i) > max:
       max = len(i)
print(list(filter(long, word)))
```

```python
def reverse(s):
    return s[::-1]
word = ['ball','bat']
print(list(map(reverse, word)))
```

```python
def ascii_value(s):
    for ch in s:
        print(ch, ord(ch))
word = input("enter a character: ")
ascii_value(word)
```

```python
def vowel(s):
    return s[0] in "aeiou"
str = ['ice','apple','pipe','clip','egg']
print(list(filter(vowel, str)))
```
